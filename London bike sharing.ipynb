```python
Q1 = df['count'].quantile(0.25)
Q3 = df['count'].quantile(0.75)
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
here are  outliers ,after removing outliers
```

```python
df_no_outlier = df[(df['count']>lower_limit)&(df['count']<upper_limit)]
df_no_outlier
```
